```python
plt.figure(figsize=(9,5))
sns.boxplot(df)
plt.xticks(rotation=45)
plt.plot()
plt.title("boxplot of student performance dataset")
plt.show()
```

```python
sns.boxplot(x="Gender", y="Final_Percentage", data=df)
plt.show()
```

```python
gb = df.groupby("Class").agg({"Class" :"count"})
plt.pie(gb['Class'] , labels = gb.index, autopct = "%1.2f%%")
plt.title("percentage of students class.")
plt.show()
```

```python
plt.figure(figsize = (4,4))
sns.barplot(x="Gender", y="Final_Percentage", data=df)
plt.show()
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x='Pass_Fail', data=df)
plt.title("Pass vs Fail Distribution")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(x='Study_Hours_Per_Day', y='Final_Percentage',
                hue='Pass_Fail', data=df)
plt.title("Study Hours vs Final Percentage")
plt.show()
```

```python
sns.histplot(df['Math_Score'], kde=True, color="blue")
plt.title("Math Score Distribution")
```

```python
sns.histplot(df['Science_Score'], kde=True, color="skyblue")
plt.title("Science Score Distribution")
```

```python
sns.histplot(df['English_Score'], kde=True, color="yellow")
plt.title("English Score Distribution")
```

```python
plt.figure(figsize=(7,5))
corr = df[['Math_Score','Science_Score','English_Score']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Scores")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x='Extracurricular_Activities', y='Final_Percentage',
                hue='Pass_Fail', data=df)
plt.title("Extracurricular Activities vs Final Percentage")
plt.show()
```
